```python
india_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "temperatures": [32, 45, 30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "temperatures": [21, 14, 35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
con=pd.concat([india_weather, us_weather], ignore_index=True)
con
```

```python
con=pd.concat([india_weather, us_weather], keys=["india", "us"])
con
```
